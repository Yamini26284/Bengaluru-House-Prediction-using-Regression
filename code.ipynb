## DATA CLEANING
 After analysing i feel that area_type, availability and society are not that important features for my cost prediction.

```python
data.info()
```

```python
data.drop(columns=['area_type','availability','society'], inplace = True) # making sure the change in dataset is amde on orinial dataset only
```

```python
data
```

```python
#dropping null rows and missing values
data.dropna(inplace=True)
data
```

```python
#distinct values for each coloum
data.location.value_counts()
```

```python
data['location']= data['location'].apply(lambda x:x.strip())
```

```python
location_stats = data.groupby('location')['location'].agg('count')
```

```python
location_stats.sort_values(ascending=False)
```

```python
#eliminating locations which have less than 11 entires, so that our ML model works accurately ignoring locations with few to no houses
location_less_than_10_entries = location_stats[location_stats<=10]
location_less_than_10_entries.sort_values(ascending=False)
```

```python
#to make our dataset more balanced we are gonna set location_less_than_10_entries to others
data['location'] = data['location'].apply(lambda x:'other' if x in location_less_than_10_entries else x)
```

```python
data['location'].value_counts()
```

```python
data
```

```python
data['size'].unique()
```

```python
# here size of the bedroom is not aligned properly, size names are distributed abnormally with diff names
# so splitting this dataset and taking numeric values
data['bedrooms']=data['size'].apply(lambda x:int(x.split(' ')[0]))
data # with the above operation we obtained a new coloumn which is more specific and direct about num of bedrooms
```

```python
data.total_sqft.unique()
```

```python
def clean(sqft):
    token = sqft.split('-')
    if len(token)==2:
        return (float(token[0])+float(token[1]))/2
    else:
        try:
            return float(sqft)
        except:
            return None
        
```

```python
data['total_sqft'] = data['total_sqft'].apply(clean)
```

```python
data.total_sqft.unique()
```

```python
data
```

```python
data.dropna(inplace=True)
data
```

```python
data.describe() #describes numeric fields
```

```python
data['sqft_per_bedroom'] = data['total_sqft']/data['bedrooms']
data.head(6)
```

```python
data2 = data[data['sqft_per_bedroom']>=300]
data2
```

```python
data2['sqft_price'] = data2['price'] * 100000 / data2['total_sqft']
data2
```

```python
data2.sqft_price.describe()
```

```python
data3 = data2[data2['sqft_price'] > 2000]
data3
```

```python
#Identifying coloumns needed for feature selection
data3.drop(columns=['size','sqft_per_bedroom','sqft_price'],axis=1,inplace=True)
```

```python
data3
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.pipeline import make_pipeline
```

```python
from sklearn.compose import make_column_transformer
```

```python
col_trans = make_column_transformer((OneHotEncoder(sparse_output=False),['location']),remainder='passthrough')
```

```python
lr = LinearRegression()
```

```python
scaler = StandardScaler()
```

```python
model = make_pipeline(col_trans,scaler, lr)
```

```python
data_input = data3.drop(columns=['price'])
data_output = data3['price']
```

```python
x_train, x_test, y_train, y_test = train_test_split(data_input,data_output,test_size=0.2) #80%- train
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
